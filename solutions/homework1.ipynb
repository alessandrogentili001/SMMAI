```python
# direct method for the solution of a linear systems 

import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import hilbert
```

```python
# RANDOM MATRIX 

relative_err_norm_2 = []
relative_err_norm_infty = []
matrix_norm_2 = []
matrix_norm_infty = []
condition_num_2 = []
condition_num_infty = []

for n in range(10,101,10):
    
    # create the random matrix
    A = np.random.rand(n,n)
    
    # create the solution 
    x_true = np.ones((n,1))
    
    # compute the right side of the sistem b = A @ x_true
    b = A @ x_true

    # compute the 2-norm of the matrix A
    A_norm_2 = np.linalg.norm(A,2)
    matrix_norm_2.append(A_norm_2)

    # compute the infinity-norm of the matrix A
    A_norm_infty = np.linalg.norm(A)
    matrix_norm_infty.append(A_norm_infty)
    
    # compute the condition number in 2-norm of the matrix A 
    A_condition_num_2 = np.linalg.cond(A,2)
    condition_num_2.append(A_condition_num_2)
    
    # compute the condition number in infinity-norm of the matrix A 
    A_condition_num_infty = np.linalg.cond(A)
    condition_num_infty.append(A_condition_num_infty)
    
    # solve the linear sistem A @ x = b
    x = np.linalg.solve(A,b)
    
    # compute the relative error in 2-norm
    x_relative_err_norm_2 = np.linalg.norm(x - x_true, 2) / np.linalg.norm(x_true, 2)
    relative_err_norm_2.append(x_relative_err_norm_2)

    # compute the relative error in infinity-norm
    x_relative_err_norm_infty = np.linalg.norm(x - x_true) / np.linalg.norm(x_true)
    relative_err_norm_infty.append(x_relative_err_norm_infty)

# plot the norm of the matrix 
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(range(10,101,10),matrix_norm_2, 'r') # plot the matrix 2-norm
plt.xlabel('dimension')
plt.ylabel('matrix norm') 
plt.legend(['2-norm'])
plt.subplot(1,2,2)
plt.plot(range(10,101,10),matrix_norm_infty, 'b') # plot the matrix 2-norm
plt.xlabel('dimension')
plt.ylabel('matrix norm') 
plt.legend(['infty-norm'])
plt.show()

# plot the condition number of the matrix 
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(range(10,101,10),condition_num_2, 'r') # plot the condition numbers in 2-norm
plt.xlabel('dimension')
plt.ylabel('condition number')
plt.legend(['2-norm'])
plt.subplot(1,2,2)
plt.plot(range(10,101,10),condition_num_infty, 'b') # plot the condition numbers in infty-norm
plt.xlabel('dimension')
plt.ylabel('condition number')
plt.legend(['infty-norm'])
plt.show()

# plot the relative error of the solution
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(range(10,101,10),relative_err_norm_2, 'r') # plot the relative error in 2-norm
plt.xlabel('dimension')
plt.ylabel('relative error')
plt.legend(['2-norm'])
plt.subplot(1,2,2)
plt.plot(range(10,101,10),relative_err_norm_infty, 'b') # plot the relative error in infty-norm
plt.xlabel('dimension')
plt.ylabel('relative error')
plt.legend(['infty-norm'])
plt.show()
```

```python
# HILBERT MATRIX 

relative_err_norm_2 = []
relative_err_norm_infty = []
matrix_norm_2 = []
matrix_norm_infty = []
condition_num_2 = []
condition_num_infty = []

for n in range(4,13):
    
    # create the Hilbert matrix
    A = hilbert(n)
    
    # create the solution 
    x_true = np.ones((n,1))
    
    # compute the right side of the sistem b = A @ x_true
    b = A @ x_true

    # compute the 2-norm of the matrix A
    A_norm_2 = np.linalg.norm(A,2)
    matrix_norm_2.append(A_norm_2)

    # compute the infinity-norm of the matrix A
    A_norm_infty = np.linalg.norm(A)
    matrix_norm_infty.append(A_norm_infty)
    
    # compute the condition number in 2-norm of the matrix A 
    A_condition_num_2 = np.linalg.cond(A,2)
    condition_num_2.append(A_condition_num_2)
    
    # compute the condition number in infinity-norm of the matrix A 
    A_condition_num_infty = np.linalg.cond(A)
    condition_num_infty.append(A_condition_num_infty)
    
    # solve the linear sistem A @ x = b
    x = np.linalg.solve(A,b)
    
    # compute the relative error in 2-norm
    x_relative_err_norm_2 = np.linalg.norm(x - x_true, 2) / np.linalg.norm(x_true, 2)
    relative_err_norm_2.append(x_relative_err_norm_2)

    # compute the relative error in infinity-norm
    x_relative_err_norm_infty = np.linalg.norm(x - x_true) / np.linalg.norm(x_true)
    relative_err_norm_infty.append(x_relative_err_norm_infty)
    
# plot the norm of the matrix 
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(range(4,13),matrix_norm_2, 'r') # plot the matrix 2-norm
plt.xlabel('dimension')
plt.ylabel('matrix norm') 
plt.legend(['2-norm'])
plt.subplot(1,2,2)
plt.plot(range(4,13),matrix_norm_infty, 'b') # plot the matrix 2-norm
plt.xlabel('dimension')
plt.ylabel('matrix norm') 
plt.legend(['infty-norm'])
plt.show()

# plot the condition number of the matrix 
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(range(4,13),condition_num_2, 'r') # plot the condition numbers in 2-norm
plt.xlabel('dimension')
plt.ylabel('condition number')
plt.legend(['2-norm'])
plt.subplot(1,2,2)
plt.plot(range(4,13),condition_num_infty, 'b') # plot the condition numbers in infty-norm
plt.xlabel('dimension')
plt.ylabel('condition number')
plt.legend(['infty-norm'])
plt.show()

# plot the relative error of the solution
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(range(4,13),relative_err_norm_2, 'r') # plot the relative error in 2-norm
plt.xlabel('dimension')
plt.ylabel('relative error')
plt.legend(['2-norm'])
plt.subplot(1,2,2)
plt.plot(range(4,13),relative_err_norm_infty, 'b') # plot the relative error in infty-norm
plt.xlabel('dimension')
plt.ylabel('relative error')
plt.legend(['infty-norm'])
plt.show()
```

```python
# compute the smallest floating point number, called epsilon

epsilon = 1
count=0
while epsilon > 0:
    if epsilon / 2 == 0:
        break
    epsilon/=2
    count+=1
    
print(f"after {count} iterations, the value of epsilon is: {epsilon}.")
```

```python
# compute the value of the Euler costant e by using the definition with the limit 
from math import exp

def e(n):
    return (1 + 1/n)**n

distances = []
for n in range(1,1001,50):
    e_n = e(n)
    distance = abs(exp(1) - e_n)
    distances.append(distance)
    
plt.plot(distances)
plt.xlabel('n')
plt.ylabel('distance')
plt.title('|e - e_approx|')
plt.show()
```

```python
# compute the rank of two matrices and their eigenvalues

import numpy as np

A = np.array([[4, 2], [1, 3]])
print(f"the matrix A is:\n {A}")
print(f"the rank of the matrix A is:\n {np.linalg.matrix_rank(A)}")
print(f"the eigenvalues of the matrix A are:\n {np.linalg.eig(A)[0]}")

B = np.array([[4, 2], [2, 1]])
print(f"the matrix B is:\n {B}")
print(f"the rank of the matrix B is: {np.linalg.matrix_rank(B)}")
print(f"the eigenvalues of the matrix B are:\n {np.linalg.eig(B)[0]}")

# from this we can conclude (also we can prove) that: a square matrix is singular iff has 0 as eigenvalue
```
